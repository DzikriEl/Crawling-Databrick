```
# File location and type
file_location = "/FileStore/tables/Mydatacrawling.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "false"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
df.printSchema()
```

```
# Create a view or table

temp_table_name = "mydatacrawling_csv"

df.createOrReplaceTempView(temp_table_name)
```

```
%sql

select * from mydatacrawling_csv
```

```
from pyspark.sql.functions import col
df = df.withColumn('_c0', col('_c0').cast('float'))
df = df.withColumn('_c1', col('_c1').cast('float'))
df = df.withColumn('_c2', col('_c2').cast('float'))
```

```
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(
inputCols=["_c0", "_c1", "_c2"],
outputCol="features")
```

```
Mydatacrawling = assembler.transform(df)
```
